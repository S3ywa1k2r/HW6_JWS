```python
import csv
import random
import matplotlib.pyplot as plt

def generate_rdnumbers(n):
    random_numbers=[random.randint(1,6) for _ in range(n)]
    return random_numbers

def main():
    result_100 = generate_rdnumbers(100)
    result_1000= generate_rdnumbers(1000)
    result_10000 = generate_rdnumbers(10000)
    result_100000 = generate_rdnumbers(100000)
    
    with open('rdnum.csv', 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['100','1000','10000','100000'])
        for i in range(0,100,1):
            temp=[]
            temp.append(result_100[i])
            temp.append(result_1000[i])
            temp.append(result_10000[i])
            temp.append(result_100000[i])
            writer.writerow(temp)
        for i in range(100,1000,1):
            temp=[]
            temp.append('')
            temp.append(result_1000[i])
            temp.append(result_10000[i])
            temp.append(result_100000[i])
            writer.writerow(temp)
        for i in range(1000,10000,1):
            temp=[]
            temp.append('')
            temp.append('')
            temp.append(result_10000[i])
            temp.append(result_100000[i])
            writer.writerow(temp)
        for i in range(10000,100000,1):
            temp=[]
            temp.append('')
            temp.append('')
            temp.append('')
            temp.append(result_100000[i])
            writer.writerow(temp)

    f=open('rdnum.csv', 'r', encoding='utf-8')
    data=csv.reader(f)
    next(data)
    result_csv_100=[]
    result_csv_1000=[]
    result_csv_10000=[]
    result_csv_100000=[]
    line=1
    for row in data:
        if line>= 1 and line<=100:
            result_csv_100.append(int(row[0]))
        if line>= 1 and line<=1000:
            result_csv_1000.append(int(row[1]))
        if line>= 1 and line<=10000:
            result_csv_10000.append(int(row[2]))
        result_csv_100000.append(int(row[3]))
        line=line+1
    
    plt.subplot(2,2,1)
    plt.title('Roll a Dice for 100 times')
    plt.hist(result_csv_100,label='100', color='r', alpha=0.5,width=0.4,bins=[0.8, 1.8, 2.8, 3.8, 4.8, 5.8, 6.8])
    plt.xticks(range(1,7,1))
    plt.xlabel('Number')
    plt.ylabel('Frequency')
    
    plt.subplot(2,2,2)
    plt.title('Roll a Dice for 1000 times')
    plt.hist(result_csv_1000,label='1000',color='b',alpha=0.5,width=0.4,bins=[0.8, 1.8, 2.8, 3.8, 4.8, 5.8, 6.8])
    plt.xticks(range(1,7,1))
    plt.xlabel('Number')
    plt.ylabel('Frequency')
    
    plt.subplot(2,2,3)
    plt.title('Roll a Dice for 10000 times')
    plt.hist(result_csv_10000,label='10000', color='g',alpha=0.5,width=0.4,bins=[0.8, 1.8, 2.8, 3.8, 4.8, 5.8, 6.8])
    plt.xticks(range(1,7,1))
    plt.xlabel('Number')
    plt.ylabel('Frequency')
    
    plt.subplot(2,2,4)
    plt.title('Roll a Dice for 100000 times')
    plt.hist(result_csv_100000,label='100000', color='y',alpha=0.5,width=0.4,bins=[0.8, 1.8, 2.8, 3.8, 4.8, 5.8, 6.8])
    plt.xticks(range(1,7,1))
    plt.xlabel('Number')
    plt.ylabel('Frequency')
    plt.tight_layout()
    
    plt.show()
    f.close()
    
    print('주사위 굴리기는 시행 횟수가 커질수록 1~6숫자가 각각 나오는 횟수가 점점 같아지는 방향으로 수렴한다.')

if __name__ =='__main__':
    main()
```
