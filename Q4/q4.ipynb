```python
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
    
def main():
    f=open('subway.csv', 'r', encoding='ANSI')
    data=csv.reader(f)
    header = next(data)
    next(data)
    subway_in=dict()
    subway_out=dict()
    subway_total=dict()
    for row in data:
        if row[3] in list(subway_in.keys()):
            subway_in[row[3]] =subway_in[row[3]]+int(row[10])+ int(row[12])
        else:
            subway_in[row[3]] = int(row[10])+ int(row[12])
            
        if row[3] in list(subway_out.keys()):
            subway_out[row[3]] =subway_out[row[3]]+int(row[11])+ int(row[13])
        else:
            subway_out[row[3]] = int(row[11])+ int(row[13])
            
        if row[3] in list(subway_total.keys()):
            subway_total[row[3]] =subway_total[row[3]]+int(row[10])+int(row[11])+int(row[12])+int(row[13])
        else:
            subway_total[row[3]] = int(row[10])+int(row[11])+int(row[12])+int(row[13])
    
    top_in = dict(sorted(subway_in.items(), key=lambda x: x[1], reverse=True)[:30])
    top_out = dict(sorted(subway_out.items(), key=lambda x: x[1], reverse=True)[:30])
    top_total=dict(sorted(subway_total.items(), key=lambda x: x[1], reverse=True)[:30])
    top_in_station=list(top_in.keys())
    top_in_passengers=list(top_in.values())
    top_out_station=list(top_out.keys())
    top_out_passengers=list(top_out.values())
    top_total_station=list(top_total.keys())
    top_total_passengers=list(top_total.values())
    font_path = "C:/Windows/Fonts/NGULIM.TTF"
    font = font_manager.FontProperties(fname=font_path).get_name()
    rc('font', family=font)
    plt.subplot(3,1,1)
    plt.title('   30')    
    plt.grid(True)
    plt.ylabel('()')
    plt.xticks(range(len(top_in_station)),top_in_station,rotation='vertical')
    plt.yticks(range(0,500001,100000))
    plt.bar(top_in_station,top_in_passengers,color='y',alpha=0.5)
    
    
    plt.subplot(3,1,2)
    plt.title('   30') 
    plt.grid(True)
    plt.ylabel('()')
    plt.xticks(range(len(top_out_station)),top_out_station,rotation='vertical')
    plt.yticks(range(0,700001,200000))
    plt.bar(top_out_station,top_out_passengers,color='b',alpha=0.5)
    
    
    plt.subplot(3,1,3)
    plt.title('   30')
    plt.grid(True)
    plt.ylabel('()')
    plt.xticks(range(len(top_total_station)),top_total_station,rotation='vertical')
    plt.yticks(range(0,1000001,200000))
    plt.bar(top_total_station,top_total_passengers,color='r',alpha=0.5)
    
    plt.gcf().set_size_inches(10, 10)
    plt.tight_layout()
    plt.show()
    f.close()
    print('       ,     ,       . ')


if __name__ =='__main__':
    main()
```
