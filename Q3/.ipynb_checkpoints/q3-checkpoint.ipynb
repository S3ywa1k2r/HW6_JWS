```python
import csv
import matplotlib.pyplot as plt
    
def main():
    f=open('jeju.csv', 'r', encoding='ANSI')
    data=csv.reader(f)
    next(data)

    man=[]
    woman=[]
    for row in data:
        for i in range(0,4,1):
            man.append(int(row[i*8+3].replace(',','')))
            woman.append(int(row[i*8+4].replace(',','')))
    year=list(range(2008,2015,2))
    bar_width=0.3
    p1=plt.bar(year,man,color='b',width=bar_width,label='Male', alpha=0.5)
    p2=plt.bar([x+bar_width for x in year],woman,width=bar_width,color='r',label='Female',alpha=0.5)
    plt.title('Male and Female population of Jeju by year')
    plt.subplots_adjust(top=1.2)
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.legend((p1[0],p2[0]),('Male','Female'))
    plt.grid(True)
    plt.xticks(range(2008,2015,2))
    plt.show()
    f.close()
    print('남/녀 인구 수는 거의 비슷하거나 남자가 조금 더 많다. 결론적으로, 제주도에는 여성의 비율이 더 높다는 속설은 거짓이다.')

if __name__ =='__main__':
    main()
```
